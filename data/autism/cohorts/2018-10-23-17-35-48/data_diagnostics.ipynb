```python
import csv
```

```python
##Create a dictionary with {mrn:diagnostic class}
with open('padded_ehrs.csv') as f:
    rd = csv.reader(f)
    padded_seq = []
    #next(rd)
    for r in rd:
        padded_seq.append(list(map(int, r)))

with open('ordered_mrns.csv') as f:
    rd = csv.reader(f)
    ehrs = {}
    ordered_mrn = []
    #next(rd)
    for idx, r in enumerate(rd):
        ordered_mrn.append(r[0])
        ehrs[r[0]] = set(padded_seq[idx])

with open('cohort-vocab.csv') as f:
    rd = csv.reader(f)
    vocab = {}
    next(rd)
    for r in rd:
        vocab[r[0]] = int(r[1])

with open('mt_to_ix.csv') as f:
    rd = csv.reader(f)
    new_vocab = {}
    #next(rd)
    for r in rd:
        new_vocab[r[0]] = int(r[1])

##OTH = 0; ASD = 1; MM = 2, AD = 3
classes = {}
for mrn in ehrs:
    classes[mrn] = set([0])
        
diseases = [['autistic', 'autism', 'pervasive', 'asperger'], ['multiple myeloma'], ['alzheimer']]
vocab_diseases = []

for d in diseases:
    tmp = []
    for n in d:
        for v in vocab:
            if v.lower().find(n)!=-1:
                tmp.append(new_vocab[str(vocab[v])])
    vocab_diseases.append(tmp)
    
for idx, vdl in enumerate(vocab_diseases):
    for vd in vdl:
        for mrn in ordered_mrn:
            if vd in ehrs[mrn]:
                    classes[mrn].add(idx+1)

for mrn in classes:
    if len(classes[mrn]) > 1:
        classes[mrn].remove(0)
```

```python
counts = {'0':0, '1':0, '2':0, '3':0}
for mrn in classes:
#     if len(classes[mrn])>1:
#         print(mrn, classes[mrn])
    for i in classes[mrn]:
        counts[str(i)] += 1

print("OTH (no overlapping): {0} -- ASD: {1} -- MM: {2} -- AD: {3}".format(counts['0'], counts['1'], counts['2'], counts['3']))
```

```python
with open('mrn_classes.csv', 'w') as f:
    wr = csv.writer(f)
    wr.writerow(['MRN', 'DISEASE_CLASS'])
    for m, c in classes.items():
        wr.writerow([m] + list(c))
```
